### 3) model 구성

* sample 개수가 적기 때문에 64개의 unit을 가진 2개의 은닉층으로 작은 network를 구성해 사용할 것임

> **TIP**   
일반적으로 train data의 개수가 적을 수록 과대적합이 더 쉽게 일어나므로 작은 model을 사용하는 것이 과대적합을 피하는 방법이다

```python
from keras import models
from keras import layers

def build_model():
    model= models.Sequential()
    model.add(layers.Dense(64, activation='relu', input_shape=(train_data.shape[1],)))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))
    
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    
    return model
```

이 network의 마지막 layer는 하나의 unit을 가지고 있고 activate function이 없다
* linear layer라고 부름
     + 전형적인 scalar regression을 위한 구성이다
* activate function을 적용하면 출력 값의 범위 제한
    + ex) 마지막 layer에 sigmoid activation이 적용되면 network가 0과 1 사이의 값을 예측하도록 train 할 것
* 마지막 layer가 순수한 linear이므로 network가 어떤 범위 값이라도 자유롭게 예측하도록 train됨

> **mse(평균 제곱 오차)**   
> * 예측과 target 사이 거리의 제곱
> * regrssion 문제에서 널리 사용되는 loss 함수다

>**MAE(평균 절대 오차)**
> * 예측과 target 사이 거리의 절댓값

### 4) K-겹 검증을 사용한 훈련 검증

epochs과 learning rate 같은 매개변수들을 조정하면서 model을 평가하기 위해 validation set을 사용


data 수가 많지 않은 상황에서 validation data set은 매우 작아짐
* 따라서, validation과 train set으로 어떤 data가 선택되었는지에 따라 validation score가 크게 달라짐
* validation set의 분할에 대한 validation score의 분산이 높다


> **K-fold cross-validation**   
> data 수가 적을 때 효과적인 validation 방법
> * data를 K개의 분할, 즉 Fold로 나누고 K개의 model을 각각 만들어 K-1 분할에서는 train을 하고, 나머지 분할에서 평가하는 방법
> * model의 validation 점수는 K개의 validation score의 평균이 된다




> **WARN**   
> train data의 통계값으로 test data 전처리를했듯이 validation data도 train data의 통계값을 사용해 전처리해야 함
> * 앞의 normalization 과정이 K-fold cross-validation 안으로 들어와야 함

```python
import numpy as np

k=4

num_val_samples= len(train_data)// k
num_epochs=100
all_scores=[]
for i in range(k):
    print('처리중인 폴드 #', i)
    val_data=train_data[i*num_val_samples: (i+1)*num_val_samples]
    val_target=train_target[i*num_val_samples: (i+1)*num_val_samples]
    
    partial_train_data=np.concatenate([train_data[:i*num_val_samples],
                                       train_data[(i+1)*num_val_samples:]], axis=0)
    
    partial_train_target=np.concatenate([train_target[:i*num_val_samples],
                                          train_target[(i+1)*num_val_samples:]], axis=0)
    
    model = build_model()
    model.fit(partial_train_data,partial_train_target, epochs=num_epochs, batch_size=1, verbose=0)
    
    val_mse, val_mae=model.evaluate(val_data, val_target, verbose=0)
    all_scores.append(val_mae)
```

```python
all_scores
```

```python
np.mean(all_scores)
```

validation set이 각각의 fold마다 다르므로 확실히 validation score가 2.3에서 2.7까지 변화가 크다   
-> 따라서 mean 값인 2.6이 다른 점수보다 훨씬 신뢰할만 하다

```python
# 모든 fold에 대해 epochs의 MAE 점수 평균을 계산
num_epochs=500
all_mae_histories=[]

for i in range(k):
    print('처리중인 폴드 #', i)
    val_data=train_data[i*num_val_samples: (i+1)*num_val_samples]
    val_target=train_target[i*num_val_samples: (i+1)*num_val_samples]
    
    partial_train_data=np.concatenate([train_data[:i*num_val_samples],
                                       train_data[(i+1)*num_val_samples:]], axis=0)
    
    partial_train_target=np.concatenate([train_target[:i*num_val_samples],
                                          train_target[(i+1)*num_val_samples:]], axis=0)
    
    model=build_model()
    history=model.fit(partial_train_data, partial_train_target, validation_data=(val_data, val_target), epochs=num_epochs, batch_size=1, verbose=0)
    
    mae_history=history.history['val_mae']
    all_mae_histories.append(mae_history)
```

```python
# EPOCH 별로 FOLD의 MAE 평균값을 계산
average_mae_history=[
    np.mean([x[i] for x in all_mae_histories]) for i in range(num_epochs)]
```

그래프가 범위가 크고 변동이 심해서 보기 어려움
1. 곡선의 다른 부분과 scale이 많이 다른 첫 10개의 data point 제외
2. 부드러운 곡선을 얻기위해 각 point를 이전 point의 지수 이동 평균(exponential moving average)로 대체

```python
smooth_mae_history2=average_mae_history[10:]
```

```python
plt.plot(range(1,len(smooth_mae_history2)+1), smooth_mae_history2)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
```

>**지수 이동 평균(exponential moving average)**   
>* 시계열 data를 부드럽게 만드는 기법 중 하나
>* previous point에 factor를 곱한 것과 현재 point에 (1-factor)을 곱해 합산한 것 

```python
def smooth_curve(points, factor=0.9):
    smoothed_points=[]
    for point in points:
        if smoothed_points:
            previous=smoothed_points[-1]
            smoothed_points.append(previous*factor+point*(1-factor))
        else:
            smoothed_points.append(point)
    
    return smoothed_points
```

```python
smooth_mae_history=smooth_curve(average_mae_history[10:])
```

```python
plt.plot(range(1,len(smooth_mae_history)+1), smooth_mae_history)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
```

이 그래프를 보면 validation MAE가 90번째 epochs 이후 줄어드는 것을 멈춤
* 이 지점 이후로는 과대적합이 시작됨

# CHAPTER_4. Machine Learning의 기본 요소

## 4.1 Machine Learning의 4가지 분류

3장에서 binary classification, multi classification, scalar regression 3종류의 machine learning 문제를 다룸
* 이 셋은 모두 supervised learning의 예다
    + supervised learning의 목표는 train data의 입력과 target 사이에 있는 관계를 학습하는 것

이것은 빙산의 일각!machine learning은 4개의 큰 범주안에 속한다


### 1) supervised learning
지도 학습, 가장 흔한 경우
* sample data가 주어지면 알고있는 target( annotation이라고도 함 )에 input data를 mapping하는 방법을 학습
    + sample data는 종종 사람이 labeling한 data일 수 있다
* 광학 문자 판독, 음성 인식, 이미지 분류, 언어 번역 같은 딥러닝의 거의 모든 application이 일반적으로 이 범주에 속함


**supervised learning은 대부분 classification과 regression으로 구성되지만 특이한 변종도 있다**
1. sequence generation: 사진이 주어지면 이를 설명하는 캡션을 생성
    * sequence 생성은 일련의 분류문제로 재구성할 수 있다.
2. syntax tree 예측: 문장이 주어지면 분해된 syntax tree를 예측
3. object detection: 사진이 주어지면 사진 안의 특정 물체 주위에 bounding box를 그림
    * 이는 많은 bounding box가 주어졌을 때, box 내용을 분류하는 classification 문제로 표현 가능
    * 또는 bounding box의 좌표를 vector regression로 예측하는 regression과 classification이 결합된 문제로 표현 가능
    
4. image segmentation: 사진이 주어졌을 때 pixel 단위로 특정 물체에 masking


### 2) unsupervised learning
비지도 학습
* 어떤 target도 사용하지 않고 input data에 대한 변환 관계를 찾는다
* data 시각화, data 압축, data의 noise 제거 또는 data에 있는 상관관계를 더 잘 이해하기 위해 사용
* data 분석에서 빼놓을 수 없는 요소
* 종종 supervised learning 문제를 풀기 전에 data set을 더 잘 이해하기 위해 필수적으로 거치는 단계
* **Dimensionality reduction**과 **Clustering**이 가장 잘 알려져 있다


### 3) self-supervised learning
자기지도 학습
* 학습과정에 사람이 개입하지 않는 supervised learning
    + 자기지도 학습은 지도 학습이지만 사람이 만든 label은 사용하지 않는다
* label이 여전히 필요하지만 보통 경험적인 알고리즘(heuristic algorithm)을 사용해서 input data로부터 생성
* Autoencoder가 잘 알려진 예다
    + target은 수정하지 않은 원본 입력
    + 같은 방식으로 전 frame 또는 단어가 주어지면, 비디오의 다음 frame 또는 단어를 예측하는 것
    + 이 경우 미래의 input data로부터 지도되기 때문에 temporally supervised learning 이라 한다
* 매커니즘을 어디에 중점을 두는지에 따라 지도학습 또는 비지도 학습으로 재해석 될 수 있다

### 4) reinforcement learning
강화학습
* 강화학습에서 agent는 환경에 대한 정보를 받아 보상을 최대화하는 행동을 선택하도록 학습된다
    + ex) 강화학습으로 훈련된 신경망은 비디오게임화면을 입력으로 받고 게임 점수를 최대화하기 위한 게임 내 행동을 출력
    
* 아직 게임 외에 실질적 연구 성공 사례는 없지만, 자율주행자동차, 자원관리, 교육 등의 분야에서 기대되어지는 분야다

## 4.2 Machine Learning model 평가

과대적합의 위험성;   
3장에서 본 예제들은 몇 번의 Epochs 후에 모두 과대적합되기 시작했다
* 즉 train data의 성능에 비해 처음 본 data에 대한 성능이 좋아지지는 않음
* 반면 train data의 성능은 train할 수록 증가

machine learning의 목표는 처음 본 data에서 잘 작동하는 일반화된 model을 얻는 것; 즉 과대적합은 주요 장애물

> 이번 장의 목표: machine learning model의 성능을 어떻게 측정하는지 

### 1) Train, Validation, Test set

model 평가의 핵심은 data를 항상 train, validation, test 3개의 set으로 나누는 것이다
* train set에서 model을 훈련하고 validation에서 model을 평가, model 출시 준비가 되면 test에서 최종적으로 딱 한번 model test
* test set에서 모델을 평가하지 않는 이유 = model을 개발할 때, 항상 model의 설정을 튜닝하기 때문
    + 예를 들어 layer 수, unit의 수를 선택 -> hyperparameter
    + validation set에서 model 성능을 평가하여 튜닝을 수행
    + 본질적으로 이런 튜닝도 어떤 parameter 공간에서 좋은 설정을 찾는 train이다
    
validation set의 성능을 기반으로 model의 설정을 튜닝하면 validation set으로 model을 직접 train하지 않더라도 빠르게 validation set에 과대적합될 수 있다.

> **information leak**
> validation set의 model 성능에 기반해 model hyperparameter를 조정할 때 마다, validation 데이터에 관한 정보가 model로 새는 것을 의미
>* 하나의 parameter에 대해서 단 한번만 튜닝한다면, 아주 적은 정보가 누설 -> 이 validation set으로는 model을 평가할만 함
>* 하지만 다시 validation set을 통해 model을 조정하는 과정을 여러번 반복하면, model에 정보를 많이 노출   
>
>결국 validation data에 맞춰 최적화했기 때문에 validation data에서 의도적으로 잘 수행되는 model 생성

최적화된 validation data가 아니라 완전 새로 보는 data에 대한 성능을 알아보고 싶다면 test set을 이용
* 단, model은 간접적으로라도 test set에 대한 어떤 정보도 얻어서는 안된다
* test set 성능에 기초해 튜닝한 model의 모든 설정은 model의 일반화 능력을 왜곡시킬 것

#### 2. K-Fold cross validation

* 이 방법은 model의 성능이 data 분할에 따라 편차가 클 때 도움이 된다

* 최종점수는 모든 K-fold cross validation을 실행해서 얻은 점수의 평균이다
* 결국 P(반복횟수)xK개의 model을 훈련하고 평가하므로 비용이 많이 든다

### 2) 기억해야 할 것

1. 대표성있는 data:    
    train data와 test data가 주어진 data에 대표성이 있어야 한다
    * 숫자 img를 분류하는 문제에서 sample 배열이 class 순서대로 나열되어 있다고 가정했을 때;
        + 배열의 첫 80%를 train data로 나머지 20%를 test set으로 만들게 되면, train에는 0~7만이 담겨 있고 test에는 8~9만 담게 된다
    * train set과 test set으로 나누기 전 data를 무작위로 섞는 것이 일반적
    
2. 시간의 방향:   
    과거로부터 미래를 예측하려고 한다면 data 분할 전 무작위로 섞어서는 절대 안된다
    * 미래의 정보가 누설되기 떄문
    * model이 사실상 미래 data에서 train 될것
    * test set의 모든 data가 미래의 것이어야 함

3. data 중복:   
    한 data set에 어떤 data가 두번 등장하면, train set과 validaton set에 데이터가 중복될 수 있다
    * 결국 train data의 일부로 test를 하는 경우가 초래
    * train set과 validation set이 중복되지 않는 확인해야 함
